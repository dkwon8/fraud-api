# Credit Card Fraud Detection — EDA

Dataset: Kaggle ULB/Worldline "Credit Card Fraud Detection"
- 284,807 transactions, 31 columns
- V1–V28: PCA-transformed features (anonymized)
- Time, Amount: original features
- Class: 0 = normal, 1 = fraud

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/creditcard.csv")
print(f"Shape: {df.shape}")
df.head()
```

```python
# Should be 0 — this dataset has no missing values
print(f"Total nulls: {df.isnull().sum().sum()}")
print(f"\nDtypes:\n{df.dtypes}")
```

```python
print("Class distribution:")
print(df["Class"].value_counts())
print(f"\nFraud rate: {df['Class'].mean():.4%}")

fig, ax = plt.subplots(figsize=(6, 4))
df["Class"].value_counts().plot(kind="bar", ax=ax, color=["steelblue", "salmon"])
ax.set_title("Class Distribution")
ax.set_xticklabels(["Normal", "Fraud"], rotation=0)
ax.set_ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

df[df["Class"] == 0]["Amount"].hist(bins=50, ax=axes[0], color="steelblue", alpha=0.7)
axes[0].set_title("Amount — Normal Transactions")
axes[0].set_xlabel("Amount")

df[df["Class"] == 1]["Amount"].hist(bins=50, ax=axes[1], color="salmon", alpha=0.7)
axes[1].set_title("Amount — Fraud Transactions")
axes[1].set_xlabel("Amount")

plt.tight_layout()
plt.show()

print(f"Normal — mean: ${df[df['Class']==0]['Amount'].mean():.2f}, median: ${df[df['Class']==0]['Amount'].median():.2f}")
print(f"Fraud  — mean: ${df[df['Class']==1]['Amount'].mean():.2f}, median: ${df[df['Class']==1]['Amount'].median():.2f}")
```
